```python
import os
import re
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['font.family'] = 'sans-serif'
mpl.rcParams['font.sans-serif'] = 'Helvetica'
mpl.rcParams['pdf.use14corefonts'] = True
mpl.rcParams['pdf.fonttype'] = 42
%config InlineBackend.figure_format = 'retina'
```

```python
df = pd.read_csv('../01_data/TNBC23_T1histBeta.csv')
```

```python
colour_tis = {'Chest wall': '#551A8B', 'Diaphragm': '#E31A1C', 'Lung': '#1F78B4', 'Pericardium': '#33A02C', 'Lymph nodes': '#FF7F00'}
colour_tim = ['#7FFF00', '#0000FF', '#9400D3']
```

```python
plt.figure(figsize=(9,4))
fig = plt.figure(dpi=300)
sns.violinplot(x="Time", y="SubPropMut", data=df, palette=colour_tim, linecolor=colour_tim, inner='box', alpha = 0.25, dodge=True)
sns.stripplot(x="Time", y="SubPropMut", hue="Organ", palette=colour_tis, data=df, alpha = 1, jitter=True)
sns.set_context('paper')
plt.tight_layout()
plt.ylim(-0.1, 0.6)
plt.savefig('../03_outs/TNBC_ViolinPlots_bySample_MUT_SubclonalProp_19OCT23.pdf')
```

```python
plt.figure(figsize=(9,4))
fig = plt.figure(dpi=300)
sns.violinplot(x="Time", y="SubPropNeo", data=df, palette=colour_tim)
sns.stripplot(x="Time", y="SubPropNeo", hue="Organ", palette=colour_tis, data=df, alpha = 1, jitter=True)
sns.set_context('paper')
plt.tight_layout()
plt.ylim(-0.25, 0.6)
plt.savefig('../03_outs/TNBC_ViolinPlots_bySample_NEO_SubclonalProp_10OCT23.pdf')
```

```python
plt.figure(figsize=(9,4))
ax1 = sns.boxplot(x="Time", y="SubPropMut", data=df, palette=colour_tim)
plt.setp(ax1.artists,fill=False)
sns.stripplot(x="Time", y="SubPropMut", hue="Tissue", palette=colour_tis, data=df, alpha = 0.95, jitter=True)
plt.savefig('../03_outs/TNBC_BoxPlots_bySample_MUT_SubclonalProp.pdf')
```

```python
plt.figure(figsize=(9,4))
ax1 = sns.boxplot(x="Time", y="SubPropNeo", data=df, palette=colour_tim)
plt.setp(ax1.artists,fill=False)
sns.stripplot(x="Time", y="SubPropNeo", hue="Tissue", palette=colour_tis, data=df, alpha = 0.95, jitter=True)
plt.savefig('../03_outs/TNBC_BoxPlots_bySample_NEO_SubclonalProp.pdf')
```
